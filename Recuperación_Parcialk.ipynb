```python
import numpy as np
import matplotlib.pyplot as plt
import sympy  as sp
```

```python
t=sp.Symbol("t")
C=sp.Function("C")
```

Contagios diarios

```python
t=np.arange(0, 200)
k= 0.117
C=20000/(1+19999*np.exp(-k*t))
dc_dt= 0.117*C*(1-(C/20000))


plt.title("Comportamiento Pandemia")
plt.plot(C,dc_dt, color="k")
plt.plot(10000,585, "d")
plt.legend=("Max de contagidos")
plt.axhline(585,linestyle="--",color="m")
plt.axvline(10000,linestyle="--",color="m")
plt.xlabel("Poblacion suceptible al contagio")
plt.ylabel("dC/dt")
plt.annotate("Punto Max",xy= (10000,585))
plt.figure(figsize=(24,10)) #tamaño
```

Día que se acabará la pandemia

```python
def dias(C):
  if (C==10000):
    d=(np.log(19999)/0.117)
  return d
dias(10000)
```

Contagios acumulados

```python
plt.title("Pico de la Pandemia")
plt.axvline(dias(10000),linestyle="--")
plt.axhline(10000,linestyle="--")
plt.plot(C,color="m",linestyle="-.")
plt.ylabel("Contagiados ")
plt.xlabel("Dias")
plt.figure(figsize=(30,13)) #tamaño
```
